## **7.3 Animation from Scratch**

```python
l2 = latam.pivot(index = "year",
                 columns = "country",
                 values = "controlc")
l2 = l2.reset_index()
l2['year_str'] = l2['year'].astype(str)
l2.head()
```

```python
# The frame
fig = go.Figure(
    layout = go.Layout(
        updatemenus = [dict(type = 'buttons', direction = 'right', x = 0.9, y = 1.16),],
        xaxis =dict(range = ['2015', '2019'],
                    autorange = False, tickwidth = 2, dtick =1,
                    title_text = 'Year'),
        yaxis = dict(range = [0, 101],
                     autorange = False,
                     title_text = 'Controlc'),
        title = 'Percentile rank: Control of corruption in Latin America',
        title_font_size = 30,
        title_x = 0.5
    )
)
```

```python
# Add Traces

init = 1

## Colombia
fig.add_trace(
    go.Scatter(
        x = l2.year[:init],
        y = l2.Colombia[:init],
        name = 'Colombia',
        line = dict(color = 'orange'),
        mode ='lines'
    )
)

## Argentina
fig.add_trace(
    go.Scatter(
        x = l2.year[:init],
        y = l2.Argentina[:init],
        name = 'Argentina',
        line = dict(color = 'blue'),
        mode ='lines'
    )
)

## Chile
fig.add_trace(
    go.Scatter(
        x = l2.year[:init],
        y = l2.Chile[:init],
        name = 'Chile',
        line = dict(color = 'green'),
        mode ='lines'
    )
)

## Honduras
fig.add_trace(
    go.Scatter(
        x = l2.year[:init],
        y = l2.Honduras[:init],
        name = 'Honduras',
        line = dict(color = 'red'),
        mode ='lines'
    )
)
```

```python
# Frames
frames = [
    go.Frame(
        data = [
            go.Scatter(x = l2.year[:k], y = l2.Colombia[:k]),
            go.Scatter(x = l2.year[:k], y = l2.Argentina[:k]),
            go.Scatter(x = l2.year[:k], y = l2.Chile[:k]),
            go.Scatter(x = l2.year[:k], y = l2.Honduras[:k]),
        ]
    )
    for k in range(init, len(l2)+1)
]

## Animation
fig.update(frames=frames)
```

```python
# Play button
fig.update_layout(
   updatemenus = [
       dict(
           buttons =list([
              dict(
               label = 'Play',
               method = 'animate',
               args = [None, {'frame':{'duration':400}}]
              )
           ]

           )
        )
    ]
)
```

### **7.3.1 Description**

This graph allows us to see the relationship between 4 chosen Latin American countries over time with respect to the control of corruption in the region; It is observed that the best country in this variable is Chile, which remains over time in the highest percentile between 80 and 100, then there are Argentina and Colombia with a very small gap where they have improved over time, especially Argentina between the 40th and 60 percentiles. Finally, we have Honduras in the lowest percentile between 20 and 40, indicating that they have high corruption rates and it continues to decline in recent years.
